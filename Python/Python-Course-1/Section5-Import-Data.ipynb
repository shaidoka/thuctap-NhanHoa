```python
import datetime
from datetime import timedelta
import calendar
```

```python
now = datetime.datetime.now()

# Calculate the number of days to subtract to get to the last Saturday
subtract_days = (now.weekday() - 5) % 7
print(subtract_days)

# Subtract the necessary days
last_saturday = now - datetime.timedelta(days=subtract_days)

# Print the date of last Saturday
print(last_saturday.day)
```

```python
# Get datimetime
now = datetime.datetime.now()
month = now.month
day = now.day
year = now.year
first_day, num_days = calendar.monthrange(year, month)

# Calculate the number of saturday of this month
num_saturdays = (num_days + first_day - 5) //7 + 1

# Calculate the date of last saturday
last_saturday = now - timedelta(days=(now.weekday() - 5) % 7)

# Calculate what is the number of month's saturday of the last saturday
saturday_of_month = (last_saturday.day - 1) // 7 + 1

# Calculate the number of saturday of last saturday's month
last_first_day, last_num_days = calendar.monthrange(last_saturday.year, last_saturday.month)
last_num_saturdays = (last_num_days + last_first_day - 5) //7 + 1
```

```python
print(saturday_of_month) 
print(last_num_saturdays)
```
