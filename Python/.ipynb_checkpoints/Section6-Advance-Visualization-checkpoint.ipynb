```python
# Stacked Histogram
```

```python
sns.set_style("darkgrid")
plt.hist(movies[movies.Genre == "Action"].BudgetMilions, bins=15)
plt.hist(movies[movies.Genre == "Drama"].BudgetMilions, bins=15)
plt.hist(movies[movies.Genre == "Thriller"].BudgetMilions, bins=15)
plt.show()
```

```python
plt.hist([movies[movies.Genre == "Action"].BudgetMilions, movies[movies.Genre == "Drama"].BudgetMilions], bins=15, stacked=True)
plt.show()
```

```python
plt.hist([movies[movies.Genre == "Action"].BudgetMilions, movies[movies.Genre == "Drama"].BudgetMilions], bins=15)
plt.show()
```

```python
list1 = list()
mylabels = list()
for gen in movies.Genre.cat.categories:
    list1.append(movies[movies.Genre == gen].BudgetMilions)
    mylabels.append(gen)

h = plt.hist(list1, bins=15, stacked=True, label=mylabels)
plt.legend()
plt.show()
```

---

```python
# Controlling Axes and Adding Diagonals
g = sns.FacetGrid(movies, row='Genre', col='Year', hue='Genre')
kws = dict(s=50, linewidth=0.5, edgecolor='Black')
g = g.map(plt.scatter, 'CriticRating', 'AudienceRating', **kws)
g.set(xlim=(0,100), ylim=(0,100))

for ax in g.axes.flat:
    ax.plot((0,100), (0,100), c="gray", ls="--")

g.add_legend()
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
import seaborn as sns
plt.rcParams['figure.figsize'] = 15,15
```

```python
sns.set_style("darkgrid", {"axes.facecolor": "gray"})
f, axes = plt.subplots(2, 2)
k1 = sns.kdeplot(x = movies.BudgetMilions, y = movies.AudienceRating, \
                 shade=True, shade_lowest=True, cmap='coolwarm', \
                 ax=axes[0,0])
k1b = sns.kdeplot(x = movies.BudgetMilions, y = movies.AudienceRating, \
                 cmap='Greens', ax=axes[0,0])

k2 = sns.kdeplot(x = movies.BudgetMilions, y = movies.CriticRating, \
                 shade=True, shade_lowest=True, cmap='inferno', \
                 ax=axes[0,1])
k1b = sns.kdeplot(x = movies.BudgetMilions, y = movies.CriticRating, \
                 cmap='cool', ax=axes[0,1])

k3 = sns.violinplot(data=movies[movies.Genre=='Drama'], \
                  x='Year', y='CriticRating', ax=axes[1,0])
k4 = sns.kdeplot(x=movies.CriticRating, y=movies.AudienceRating, \
                shade=True, shade_lowest=False, cmap='Reds', \
                ax=axes[1,1])
k4b = sns.kdeplot(x=movies.CriticRating, y=movies.AudienceRating, \
                 cmap='Reds', ax=axes[1,1])
```
