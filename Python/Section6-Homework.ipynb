```python
movies.Genre = movies.Genre.astype('category')
```

```python
Filter_Genre = (movies.Genre == "action") | (movies.Genre == "comedy") | (movies.Genre == "adventure") | (movies.Genre == "animation") | (movies.Genre == "drama")
```

```python
movies.columns
```

```python
movies.columns = ['WeekDay', 'Director', 'Genre', 'Movie', 'ReleaseDate',
       'Studio', 'AdjustedGross', 'Budget', 'Gross',
       'ImdbRating', 'MovieLensRating', 'Overseas', 'OverseasPercent',
       'Profit', 'ProfitPercent', 'Runtime', 'US',
       'GrossPercentUS']
```

```python
movies2 = movies[Filter_Genre]
```

```python
movies2
```

```python
# Filter the dataframe by studio
movies3 = movies2[(movies2.Studio == 'Buena Vista Studios') | (movies2.Studio == 'Fox') | (movies2.Studio == 'Paramount Pictures') | (movies2.Studio == 'Sony') | (movies2.Studio == 'Universal') | (movies2.Studio == 'WB')]
```

```python
fig1 = sns.boxplot(data=movies3, x=movies3.Genre, y=movies3.GrossPercentUS, order=movies3.Genre.unique(), \
            color="#dfdde5", showfliers=False)

sns.stripplot(x='Genre', y='GrossPercentUS', data=movies3, jitter=True, \
              size=6, linewidth=0, hue = 'Studio', alpha=0.7, 
             order=movies3.Genre.unique())

fig1.axes.set_title('Domestic Gross % by Genre',fontsize=30)
fig1.set_xlabel('Genre',fontsize=20)
fig1.set_ylabel('Gross % US',fontsize=20)

# Define where to place the legend
fig1.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```
